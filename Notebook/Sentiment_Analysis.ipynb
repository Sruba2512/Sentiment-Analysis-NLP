```python
# To import the required libraries
import pandas as pd ## For Data Manipulation and Analysis;
import requests ## For sending HTTP requests;
import re ## For working with Regular Expressions;
import os ## For interacting with the Operating System;

## For Web Scraping framework;
from bs4 import BeautifulSoup
from bs4.element import Comment
```

```python
# To read the required file
data = pd.read_excel('Input.xlsx')
print(data) ## Displays the specified data;
```

```python
# To define a function for removing punctuations
def preprocessor(text):
    ## Defining conditional statements;
    if isinstance((text), (str)): ## Checks if the 'text' is of 'string' type;
        text = re.sub('<[^>]*>', ' ', text) ## Searches for the pattern and replaces;
        text = re.sub('[\W]+', ' ', text.lower()) ## Searches for the pattern and replaces;
        return text
    if isinstance((text), (list)): ## Checks if the 'text' is of specified type;
        return_list = [] ## Creates an empty list;
        for i in range(len(text)): ## 'i' is the loop control variable;
            temp_text = re.sub('<[^>]*', '', text[i]) ## Searches for the pattern and replaces;
            temp_text = re.sub('[\W]+', '', temp_text.lower()) ## Searches for the pattern and replaces;
            return_list.append(temp_text) ## Appends specified element to the list;
        return (return_list)
```

```python
url_ids = [] ## Creates an empty list;
url_list = [] ## Creates an empty list;
titles = [] ## Creates an empty list;
texts = [] ## Creates an empty list;

for url_id,url in zip(data['URL_ID'], data['URL']):
    ## Following user-agent string belongs to Opera browser running on Windows 10;
    my_headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36' +
                  ' (KHTML, like Gecko) Chrome/61.0.3163.100Safari/537.36'}

    page = requests.get(url, headers = my_headers) ## Sends a GET request to the specified url;

    if page.status_code == 200:
        url_ids.append(url_id) ## Appends specified element to the list;
        url_list.append(url) ## Appends specified element to the list;
        soup_packtpage = BeautifulSoup(page.text) ## Creates a 'BeautifulSoup' object;
        title = soup_packtpage.find('title').text
        title = re.sub(' +', ' ', title) ## Searches for the pattern and replaces;
        titles.append(title) ## Appends specified element to the list;
        text = soup_packtpage.find("div", class_="td-post-content").text
        text = re.sub('\n', ' ', text) ## Searches for the pattern and replaces;
        text = re.sub(' +', ' ', text) ## Searches for the pattern and replaces;
        texts.append(text) ## Appends specified element to the list;
```

* <font size = "4"> -- Cleaning using Stop Words Lists -- </font>
---


```python
# To set the path
path = '/gdrive/MyDrive/Blackcoffer Assignment/20211030 Test Assignment/StopWords'
os.chdir(path) ## Changes the current working directory to the specified path;

stop_words_list = [] ## Creates an empty list;

## Checks in the list of all files and directories;
for file in os.listdir():
    if file.endswith('.txt'): ## Checks if the string ends with the specified value;
        file_path = f'{path}/{file}'
        ## Opens the file with specific encoding format;
        f = open(file_path, 'r', encoding = "ISO-8859-1")
        data = f.readlines() ## Returns a list containing each line of the file;
        f.close() ## Closes the file format;
        if file.endswith('Currencies.txt'):
            lines = [] ## Creates an empty list;
            for line in data:
                line = line.rstrip() ## Removes any trailing characters;
                line = line.split('|') ## Splits the string with '|';
                lines = lines + line
        else:
            lines = [line.rstrip() for line in data]
        stop_words_list = stop_words_list + lines

stop_words = set(preprocessor(stop_words_list)) ## Function Calling;
print(stop_words)
```

* <font size = "4"> -- Creating a dictionary of Positive and Negative words -- </font>
---

```python
# To create a dictionary of Positive words

## Opens the required file with specific encoding format;
file_p = open('/gdrive/MyDrive/Blackcoffer Assignment/20211030 Test Assignment/MasterDictionary/positive-words.txt', 'r', encoding = "ISO-8859-1")
data_p = file_p.readlines() ## Returns a list containing each line of the file;
file_p.close() ## Closes the file format;

positive_words = [line.rstrip() for line in data_p if len(line.rstrip()) > 1]
positive_dict = set(positive_words)
print(positive_dict) ## Displays the positive words;
```

```python
# To create a dictionary of Negative words

## Opens the required file with specific encoding format;
file_n = open('/gdrive/MyDrive/Blackcoffer Assignment/20211030 Test Assignment/MasterDictionary/negative-words.txt', 'r', encoding = "ISO-8859-1")
data_n = file_n.readlines() ## Returns a list containing each line of the file;
file_n.close() ## Closes the file format;

negative_words = [line.rstrip() for line in data_n if len(line.rstrip()) > 1]
negative_dict = set(negative_words)
print(negative_dict) ## Displays the Negative words;
```

* <font size = "4"> -- Extracting Derived variables -- </font>
* <font size = "4"> -- Analysis of Readability -- </font>
* <font size = "4">-- Average Number of Words Per Sentence -- </font>
* <font size = "4"> -- Complex Word Count -- </font>
*	<font size = "4"> -- Word Count -- </font>
*	<font size = "4"> -- Syllable Count Per Word -- </font>
*	<font size = "4"> -- Personal Pronouns -- </font>
*	<font size = "4"> -- Average Word Length -- </font>
---

```python
Positive_Score = [] ## Creates an empty list;
Negative_Score = [] ## Creates an empty list;
Polarity_Score = [] ## Creates an empty list;
Subjectivity_Score = [] ## Creates an empty list;

Average_Sentence_Length = [] ## Creates an empty list;
Percentage_of_Complex_words = [] ## Creates an empty list;
Fog_Index = [] ## Creates an empty list;

Average_Number_of_Words_Per_Sentence = [] ## Creates an empty list;
Complex_Word_Count = [] ## Creates an empty list;
Word_Count = [] ## Creates an empty list;

Syllable_Count =[] ## Creates an empty list;
Personal_Pronouns = [] ## Creates an empty list;
Average_Word_Length = [] ## Creates an empty list;

# To define a function for counting Syllable per word
def count_syllables(word):
    return len(
        re.findall('(?!e$)[aeiouy]+', word, re.I) +
        re.findall('^[^aeiouy]*e$', word, re.I)
    )

## Compiles a regular expression pattern;
pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b', re.I)

for text in texts:
    pronouns = pronounRegex.findall(text)
    Personal_Pronouns.append(len(pronouns)) ## Appends specified element to the list;
    sentences = text.split('.') ## Splits the string with '.';
    processed_text = preprocessor(text) ## Function Calling;
    words = processed_text.split() ## Splits the string;

    positive_score = 0 ## Initializes the variable;
    negative_score = 0 ## Initializes the variable;
    cleaned_words = 0 ## Initializes the variable;
    complex_words = 0 ## Initializes the variable;
    word_chars = 0 ## Initializes the variable;
    syll_words = 0 ## Initializes the variable;

    for word in words:
        if word not in stop_words: ## Condition checking;
            word_chars = word_chars + len(word)
            cleaned_words = cleaned_words + 1
            if word in positive_dict: ## Condition checking;
                positive_score = positive_score + 1
            elif word in negative_dict: ## Condition checking;
                negative_score = negative_score + 1
            if count_syllables(word) > 2: ## Condition checking;
                complex_words = complex_words + 1
            if count_syllables(word) >= 1: ## Condition checking;
                syll_words = syll_words + 1

    Positive_Score.append(positive_score) ## Appends specified element to the list;
    Negative_Score.append(negative_score) ## Appends specified element to the list;
    Polarity_Score.append((positive_score - negative_score)/(positive_score + negative_score + 0.000001))
    Subjectivity_Score.append((positive_score + negative_score)/(cleaned_words + 0.000001))

    average_sentence_length = cleaned_words/len(sentences) ## Applying formula;
    Average_Sentence_Length .append(average_sentence_length)

    per_complex_words = complex_words/cleaned_words ## Applying formula;
    Percentage_of_Complex_words.append(per_complex_words)

    Fog_Index.append(0.4*(average_sentence_length + per_complex_words))

    Average_Number_of_Words_Per_Sentence.append(cleaned_words/len(sentences))
    Complex_Word_Count.append(complex_words) ## Appends specified element to the list;
    Word_Count.append(cleaned_words) ## Appends specified element to the list;

    Syllable_Count.append(syll_words/1042) ## Appends specified element to the list;
    Average_Word_Length.append(word_chars/cleaned_words)
```

```python
df = pd.DataFrame({'URL_ID':url_ids, 'URL': url_list,'POSITIVITY_SCORE':Positive_Score,'NEGATIVITY_SCORE':Negative_Score,
                   'POLARITY_SCORE':Polarity_Score,'SUBJECTIVITY_SCORE':Subjectivity_Score,'AVG_SENTENCE_LENGTH': Average_Sentence_Length,
                   'PERCENTAGE_OF_COMPLEX_WORDS':Percentage_of_Complex_words,'FOG_INDEX':Fog_Index,'AVG_NUMBER_OF_WORDS_PER_SENTENCE':Average_Number_of_Words_Per_Sentence,
                   'COMPLEX_WORD_COUNT':Complex_Word_Count,'WORD_COUNT':Word_Count,'Syllable_Count':Syllable_Count,'PERSONAL_PRONOUNS':Personal_Pronouns,'AVG_WORD_LENGTH':Average_Word_Length})
df
```

```python
from IPython.display import HTML
import base64

def create_download_link(df, title = "Download CSV file", filename = "data.csv"):
    csv = df.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

## Let's create a link for CSV file
create_download_link(df)
```
